```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\Jupyter\CarPrice_Assignment (1).csv")

df['carCompany'] = df['CarName'].str.split(' ').str[0].str.lower()


df['carCompany'] = df['carCompany'].replace({
    'vw': 'volkswagen',
    'vokswagen': 'volkswagen',
    'porcshce': 'porsche',
    'toyouta': 'toyota',
    'maxda': 'mazda'
})
d_carcompany = pd.get_dummies(df['carCompany'], drop_first=True)
d_cylno = pd.get_dummies(df['cylindernumber'],drop_first=True)
d_engntyp = pd.get_dummies(df['enginetype'],drop_first=True)
d_drivewheel = pd.get_dummies(df['drivewheel'],drop_first=True)
d_carbody = pd.get_dummies(df['carbody'],drop_first=True)
d_fuelsystem = pd.get_dummies(df['fuelsystem'],drop_first=True)

df = pd.concat([df,d_carcompany,d_cylno,d_engntyp,d_drivewheel,d_carbody,d_fuelsystem],axis=1)
```

```python
#New derived column car stability is defined above car_stability = wheelbase/length_of_car
df['car_stability'] = df['wheelbase'] / df['carlength']
print(df['car_stability'])


```

```python
#Dropping related independent variables.
df.drop(['wheelbase', 'carlength'], axis=1, inplace=True)
```

```python
#Highly co-related variables, we can get it from heatmap
import seaborn as sb
import matplotlib.pyplot as plt
import seaborn as sb
import matplotlib.pyplot as plt
plt.figure(figsize =(14,10))
numeric_df = df.select_dtypes(include=['number'])
sb.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Highly co-related variables")
plt.show()
```

```python
# Drop highly correlated and redundant columns
df.drop(['curbweight',  'carwidth', 'highwaympg', 'car_stability'], axis=1, inplace=True)
```
